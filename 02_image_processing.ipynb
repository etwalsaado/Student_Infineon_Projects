```python
# 02_image_processing.ipynb - skeleton + demo image created in code
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Create a synthetic image with a "defect" line
img = np.full((400,600,3), 200, dtype=np.uint8)  # light gray background
cv2.rectangle(img, (100,100), (500,300), (160,160,160), -1)  # object
cv2.line(img, (150,120), (450,280), (0,0,0), thickness=6)     # scratch line (defect)
cv2_imshow(img)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

# Threshold (Otsu)
ret, th = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
print("Otsu threshold:", ret)
cv2_imshow(th)

# Edge detection
edges = cv2.Canny(gray, 50, 150)
cv2_imshow(edges)
```
